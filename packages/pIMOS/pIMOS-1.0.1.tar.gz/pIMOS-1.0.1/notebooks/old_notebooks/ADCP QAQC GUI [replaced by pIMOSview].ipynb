```python
"""
ADCP QAQC Gui
""" 
import tkinter as tk
from tkinter.filedialog import askopenfilename, asksaveasfilename
    
from crude_readers.RDI_ADCP_PD0 import RDI_ADCP_PD0

from zutils import plotting as zplot
import matplotlib.pyplot as plt
import numpy as np

from tkinter import * 
from matplotlib.figure import Figure 
from matplotlib.backends.backend_tkagg import (FigureCanvasTkAgg,  
NavigationToolbar2Tk) 

import matplotlib
matplotlib.use('TkAgg')

import os
import ctypes
 
ctypes.windll.shcore.SetProcessDpiAwareness(1)

# plot function is created for  
# plotting the graph in  
# tkinter window 
class qc_tool():
    
    widths=[40, 1]   
    heights=[8]  
    hspace=[1]     
    vspace=0.5       

    bottom = 6
    left = 6
    right = 8

    zl = zplot.axis_layer(widths=widths, heights=heights, hspace=hspace, vspace=vspace, bottom=bottom, left=left, right=right)
    zl.verbose = False # Reduce printouts

    def __init__():
        
        self.zl = zl
    
    def load(self):
        
        """Open a file for editing."""
        filepath = askopenfilename(
            filetypes=[("Text Files", "*.nc"), ("All Files", "*.*")]
        )
        if not filepath:
            return

        print(filepath)

        folder = os.path.split(filepath)[0]
        file   = os.path.split(filepath)[1]

        self.load_go(folder, file)
        
    def load_demo(self): 

        folder = r'C:\Users\AZulberti\O2 Marine\OneDrive - Documents\DATA\PROJECTS\19MET-0010 Sunset Beach Metocean Measurements\4_Analysis\03 Processed Data\05.03.2020_04.06.2020\SUN1\CURRENTS_PD0_files'
        file = r'Sunset Beach Metocean Deployments_SB_FT02_Sunset 1_currents.nc'

        self.load_go(folder, file)
        
    def load_go(self, folder, file): 
        
        rr = RDI_ADCP_PD0(folder, file, method='nc')
        self.rr = rr
        
        # the figure that will contain the plot 
        fig = Figure(figsize = (20, 3), 
                     dpi = 100) 
        self.fig = fig
        
        # adding the subplot 
#         ax1 = fig.add_subplot(111) 
        if True:
            ax1 = self.zl.lay(0,0, figure=fig, force_new=True, frameon=True)
            cax1 = self.zl.lay(0,1, figure=fig, force_new=True, frameon=True)
            self.ax1 = ax1 
            self.cax1 = cax1 
            
            # plotting the graph 
        #     plot1.plot(y) 
        else:
            ax1 = fig.add_subplot(111) 
            self.ax1 = ax1 
            
        # creating the Tkinter canvas 
        # containing the Matplotlib figure 
        canvas = FigureCanvasTkAgg(fig, 
                                   master = self.window)   
        

        # creating the Matplotlib toolbar 
        toolbar = NavigationToolbar2Tk(canvas, 
                                       self.window) 
        toolbar.update() 

        canvas.draw() 

        # placing the canvas on the Tkinter window 
        canvas.get_tk_widget().pack() 
        
        self.update_plot()
        
#     def dummy(self): 

#         fig = Figure(figsize = (20, 3), 
#                      dpi = 100) 
#         self.fig = fig
#         canvas = FigureCanvasTkAgg(fig, 
#                                    master = self.window)   
        
#         toolbar = NavigationToolbar2Tk(canvas, 
#                                        self.window) 
#         toolbar.update() 
        
        
#         ##
# #         ax1 = self.zl.lay(0,0, figure=fig, resize=False)
#         rect = self.zl.get_pos_norm(0, 0)
#         ax = fig.add_axes(rect, frameon=True)
        
#         ##
        
#         canvas.draw() 
#         canvas.get_tk_widget().pack() 
                
#         # placing the canvas on the Tkinter window 
#         canvas.get_tk_widget().pack() 
        
    def export(self):
        
        sel = messagebox.askyesnocancel('Export', 'Are you sure you wish to export changes?')
        if sel is None:
            print('Export cancelled.')
            return
        if sel:
            print('Exporting changes.')
            self.rr.ds.close()
            self.rr.export()
        else:
            print('Export cancelled.')
        

    def update_plot(self, cl = 0.2, replot=True):
        
#         def get_qaqc_var(rr, var_name, flag_name):
            
#             da = rr.ds[var_name].copy()
#             ind = rr.ds[flag_name].values > 0
#             da.values[ind] = np.nan
            
#             return da 
        
        
        # Seems to be the only way to clear a colorbar
        if False:
            for i in self.ax1.collections:
                try:
                    i.colorbar.remove()
                except:
                    pass
        
        self.ax1.clear()
        plt.show()
        print('Axis cleared')
        
        
#         logind = ~self.rr.ds.flag>0
        
#         da.values[self.rr.ds.flag>0] = np.nan
        
#         da = da.where(logind, qc_code)
        call_type = 'inline'
        call_type = 'subfunction'
        call_type = 'module'
        
#         if call_type == 'inline':
            
# #             da = self.rr.ds['U'].copy()
# #             ind = self.rr.ds['flag'].values > 0
# #             da.values[ind] = np.nan
#             pass

#         if call_type == 'subfunction':
            
# #             da = get_qaqc_var(self.rr, 'U', 'flag')
#             pass
        
        if call_type == 'module':
            
            da = self.rr.get_qaqc_var('U')

        if replot:
            da.plot(ax=self.ax1, vmin=-cl, vmax=cl, cmap='RdBu', cbar_ax=self.cax1)
            print('Plotting complete.')

    def qc_xlims(self):
            
        xl = self.ax1.get_xlim()
        xl = [matplotlib.dates.num2date(x).replace(tzinfo=None) for x in xl]
        xl_np64 = [np.datetime64(x) for x in xl]
        
        print('Start: {}'.format(xl[0]))
        print('End: {}'.format(xl[1]))
        
#         ds2 = self.rr.ds.sel(time=slice(xl[0], xl[1]))
#         print(ds2)
            
        # Need something better than this.
        sel = messagebox.askyesnocancel('Change QAQC code', 'Yes: code 0 [good data]; No: code 1 [bad data]')
        if sel is None:
            return
        if sel:
            qc_code = 0.
        else:
            qc_code = 1.
        
        call_type = 'inline'
        call_type = 'subfunction'
        call_type = 'module'
        
        if call_type == 'module':
            """
            This is the module method call
            """
            
            self.rr.update_qc_flag('*', 'time', xl_np64[0], xl_np64[1], qc_code)
            
            pass
        
        else:
            
            raise(Exception)
        
        self.update_plot(replot=True)

        
    def __init__(self):
        # the main Tkinter window 
        window = Tk() 
        self.window = window
        
        # setting the title  
        window.title('Plotting in Tkinter') 

        # dimensions of the main window 
        window.geometry("700x500") 

        plot_button = Button(master = window,  
                             command = self.update_plot, 
                             height = 2,  
                             width = 10, 
                             text = "Update plot") 
        plot_button.pack() # place the button in main window 
        
        qcx_button = Button(master = window,  
                             command = self.qc_xlims, 
                             height = 2,  
                             width = 10, 
                             text = "QC xlims") 
        qcx_button.pack() # place the button in main window 

        m = Menu(window, tearoff=0)
        m.add_command(label="THIS IS JUST A DEMO CONTEXT MENU")
        m.add_separator()
        m.add_command(label="Cut")
        m.add_command(label="Copy")
        m.add_command(label="Paste")
        m.add_command(label="Reload")
        m.add_separator()
        m.add_command(label="Rename")
        
        def do_popup(event):
            try:
                m.tk_popup(event.x_root, event.y_root)
            finally:
                m.grab_release()

        qcx_button.bind("<Button-3>", do_popup)
        
        self.build_ribbon()
        
        # run the gui 
        window.mainloop() 
    
    def build_ribbon(self):
        
        def donothing():
            pass
        
        custom_font=("Verdana", 20)
        
        menubar = Menu(self.window)
        filemenu = Menu(menubar, tearoff=0)
        filemenu.add_command(label="Open", command=self.load, font=custom_font)
        filemenu.add_command(label="Open demo file", command=self.load_demo, font=custom_font)
        filemenu.add_command(label="Export changes to NetCDF", command=self.export, font=custom_font)
        filemenu.add_separator()
        filemenu.add_command(label="Exit", command=self.window.quit, font=custom_font)
        menubar.add_cascade(label="File", menu=filemenu, font=custom_font)
#         menubar.config()
        
        self.window.config(menu=menubar)
        
obj = qc_tool()
```
